```python
# Import necessary library
import numpy as np
!pip install scipy
from scipy.stats import norm
```

```python
# SCENARIO 1: 
#  Given Values
mu0 = 30          # null hypothesis mean
x_bar = 27.8       # Sample mean (minutes)
sigma = 6          # Population standard deviation (minutes)
n = 64             # Sample size
alpha = 0.05       # Significance level

# Step 1: Calculate the Standard Error (SE)
SE = sigma / np.sqrt(n)
print("Standard Error (SE)=", SE)
# Step 2: Calculate the z-score
z = (x_bar - mu0) / SE
print("z-score =", z)
# Step 3: Find two-tailed p-value
p_value = 2 * (1 - norm.cdf (abs (z) ))
print ("p-value =", p_value)
# Step 4: Decision
if p_value < alpha:
    print("Reject H0: There is evidence that the process change affected ED wait times")
else:
    print("Fail to reject H0: There is insufficient evidence that the process change affected ED wait times")
#Step 5: Hypotheses
print("\nHypotheses:")
print("H0: μ = 30 (The process change did not affect ED wait times)")
print("H1: μ ≠ 30 (The process change affected ED wait times)")

# Step 6: Interpretation
print("\nInterpretation:")
print(f"The p-value of {p_value:.4f} indicates that the observed difference in mean ED wait time "
      "is unlikely to have occurred by random chance if the true mean were 30 minutes. "
      "This suggests the process change may have affected ED wait times.")
```

```python
# SCENARIO 2:  
# Given Values
mu0 = 120         # Null hypothesis mean (mmHg)
x_bar = 117.9     # Sample mean (mmHg)
sigma = 10        # Population standard deviation (mmHg)
n = 100           # Sample size
alpha = 0.05      # Significance level

# Step 1: Calculate the Standard Error (SE)
SE = sigma / np.sqrt(n)
print("Standard Error (SE) =", SE)

# Step 2: Calculate the z-score
z = (x_bar - mu0) / SE
print("z-score =", z)

# Step 3: Find two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z)))
print("p-value =", p_value)

# Step 4: Decision
if p_value < alpha:
    print("Reject H0: There is evidence that the dietary intervention affected systolic blood pressure")
else:
    print("Fail to reject H0: There is insufficient evidence that the dietary intervention affected systolic blood pressure")
    
# Step 5: Hypotheses
print("\nHypotheses:")
print("H0: μ = 120 (The dietary intervention did not affect systolic blood pressure)")
print("H1: μ ≠ 120 (The dietary intervention affected systolic blood pressure)")

# Step 6: Interpretation
print("\nInterpretation:")
print(f"The p-value of {p_value:.4f} indicates that the observed difference in mean systolic blood pressure "
      "is unlikely to have occurred by random chance if the true mean were 120 mmHg. "
      "This suggests the dietary intervention may have had an effect.")
```
